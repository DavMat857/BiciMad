# El objetivo de este notebook es mostrar como obtener información sobre los datos a utilizar
#### Con una línea nos vale para sacar información sobre localización, id de la estacion, nombre...

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('example').getOrCreate()

filename = "SituacionesEstaciones12_2020.json"

df = spark.read.json(filename)
print("Vamos a obtener las columnas\n" ,df.columns)
print("\nVamos a mostar un ejemplo de los datos a usar\n", df.take(1)) # TODO: necesario? no se ve bien
```

```python
#Vamos a seleccionar stations
stations = df.select('stations')
print("Vamos a obtener las columnas\n" ,stations.columns) # TODO: creo que no es lo que buscabamos
print("\nVamos a mostar un ejemplo de los datos a usar\n", stations.take(1))
```

```python
from pyspark.sql.functions import col

direccion = stations.select(col("stations.address")).collect()
direccion
```

# Seleccionamos los siguientes campos
 * name
 * longitude
 * latitude
 * id

```python
seleccion = df.select(col("stations.name"), col("stations.longitude"), col("stations.latitude"),col("stations.id"))
seleccion.take(1)
```

```python
datos = (seleccion
                .select(col("name"), col("latitude"), col("longitude"), col("id"))
                .collect())

# Transformar cada fila en un diccionario que contenga solo las columnas "name" y "address"
result = []
for row in datos:
    result.append({"name": row["name"], "latitude": row["latitude"], "longitude": row["longitude"], "id": row["id"]})
#Solo queremos datos de una línea pero tenemos de todas
result[0]
```

```python
linea = result[0]
linea.keys()

for i in linea.keys():
    a = [linea[i][0]]
    print(a)
```

```python

datos_usar = {}

for key, values in linea.items():
    for i, value in enumerate(values):
        datos_usar.setdefault(i, {})
        datos_usar[i][key] = value
        
print(datos_usar)
```
